#Checking bike_lakehouse.gold.dim_customers

```python
%sql
-- Check for Uniqueness of Customer Key in gold.dim_customers
-- Expectation: No results 
SELECT 
    customer_key,
    COUNT(*) AS duplicate_count
FROM bike_lakehouse.gold.dim_customers
GROUP BY customer_key
HAVING COUNT(*) > 1;
```

#Checking bike_lakehouse.gold.dim_products

```python
%sql
-- Check for Uniqueness of Product Key in gold.dim_products
-- Expectation: No results 
SELECT 
    product_key,
    COUNT(*) AS duplicate_count
FROM bike_lakehouse.gold.dim_products
GROUP BY product_key
HAVING COUNT(*) > 1;
```

#Checking bike_lakehouse.gold.fact_sales

```python
%sql
-- Check the data model connectivity between fact and dimensions
SELECT * 
FROM bike_lakehouse.gold.fact_sales f
LEFT JOIN bike_lakehouse.gold.dim_customers c
ON c.customer_key = f.customer_key
LEFT JOIN bike_lakehouse.gold.dim_products p
ON p.product_key = f.product_key
WHERE p.product_key IS NULL OR c.customer_key IS NULL  
```
